```python
#import tensorflow

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

#create features
x = np.array([-7.0, -4.0, -1.0, 2.0, 5.0, 8.0, 11.0, 14.0])

#create labels
y = np.array([3.0, 6.0, 9.0, 12.0, 15.0, 18.0, 21.0, 24.0])

#visualize it
plt.scatter(x,y)
#mehm me pt.scatter ekn blnne x and y athara thiyna prshathare so i prshtharen thami api me deka thar relationship ek hoygnne
# e prshathre sameekarnaya thami aththtama equatin ek prediction kraddi inputs walat adala output ek hoygnna
```

```python
#lets check input and output shapes

input_shape = x.shape
output_shape = y.shape

input_shape, output_shape

#mekedi aththtma wenne eka x value ekakin eka v value ekk predict krnnai hdnne , eka hondat mathk thiygnna one
```

```python
#dan thmi mekat model ekk ghna blnne
#turn our numpy arrays into tensors

x = tf.constant(x)
y = tf.constant(y)
x, y
```

```python
#input shape eki output shape eki blmu eka eka value ekt x, y arrays 2

input_shape1 = x[0].shape
output_shape1 = y[0].shape
input_shape1, output_shape1
#ethkot mekn kiynne eka input value ekakin eka output ekk predict krnwa
```

```python
#dan mekt model ek hadaddi api aththtam hdnne y = x + 10 rekhawa thiyna sameekaranayt adalwa model ekk ethko ek krnne khmd stesp:

  #1. creating a model - define the input and output layers as well as the hidden layrs of a deep learning model
  #2. compiling a model - define the loss function (in other words, the function which tells our model how wrong it is) and the otmizer (tells our model how to improve the patterns its learning) and evaluation metrics (what we can use to interpret the performance of our model)
  #3. fitting a model- letting the modeltry to find patterns between x and y(feature and labels)
```

```python
#------------------------step - 1---------------------------------------

#set random seed
tf.random.set_seed(42)

# for this model cration we use sequential api
model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(1)
])

#-----------------------step-2 compile the model-------------------------

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])

#-----------------------step - 3 : fit the model----------------------------
model.fit(x, y, epochs=5)
```

```python
#now lets check out x and y first before try to predict a value
x, y
```

```python
#lets predict a value now
model.predict(np.array([[17.0]]))

#hondata blnna meke prediction eka wardi
```

```python
#so as we can see this is not the prediction accuracy we needed , so the value should be 27 but it gives 1.94
# so how to improve our model:

#We can improve our model, by altering the steps we took to create a model

#1. Creting a model :- Here we might add more layers, increase the number of hidden units ( all called neurons) within each of the hidden layers, change the activation function of each layer.
#2. compiling a model :- Here we migth change the optimization function or perhaps the learning rate of the optimization funtion
#3. fitting a model :- here we might fit a model for more epochs (leave it training for longer) or on more data (give the model more examples to learn from).

#creating the model again

#------------------------step - 1---------------------------------------

#set random seed
tf.random.set_seed(42)

# for this model cration we use sequential api
model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(1)
])

#-----------------------step-2 compile the model-------------------------

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])

#-----------------------step - 3 : fit the model (this time we increase the epochs----------------------------
model.fit(x, y, epochs=100)
```

```python
# reminder ourselves of the data before predicting the values again
x,y
```

```python
#lets predict again

model.predict(np.array([[17.0]]))
```

```python
#so last time we have got to 29, hari nm enna one 27 , so kalin ekt wada sahenna improvement ekk thiynwa
#dan aphu ape podi model eka improve krla blmu hariytma

# me improvement ekekdi api krnne wadi pura 1 hidden layer ekk daanwa

#------------------------step - 1---------------------------------------
#set random seed
tf.random.set_seed(42)

# for this model cration we use sequential api
model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(1)
])

#-----------------------step-2 compile the model-------------------------

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])

#-----------------------step - 3 : fit the model (this time we increase the epochs----------------------------
model.fit(x, y, epochs=100)
```

```python
#lets predict again

model.predict(np.array([[17.0]]))
```

```python
#so this solution is too much so lets make the adam optimizer and lets use 0.01 learnign rate


#------------------------step - 1---------------------------------------
#set random seed
tf.random.set_seed(42)

# for this model cration we use sequential api
model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(1)
])

#-----------------------step-2 compile the model-------------------------

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
              metrics=["mae"])

#-----------------------step - 3 : fit the model (this time we increase the epochs----------------------------
model.fit(x, y, epochs=100)
```

```python
# lets predict a value now

model.predict(np.array([[17.0]]))
```
