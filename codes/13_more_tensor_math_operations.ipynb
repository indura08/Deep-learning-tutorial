```python
#tf.amth eke godak maths perators thiynwa, methnain katah krnne ethanin kiypayak
#squaring, log, square root

import tensorflow as tf

H = tf.range(1,10)
H
```

```python
#square it
tf.square(H)
```

```python
#find the sqaure root (error ekk enna puluwan tensor eke data type eka int nm , int ewa me method ekt daanna bha )
tf.math.sqrt(tf.cast(H, dtype=tf.float32))
```

```python
#find the log (meketh ehmai int ewa bha float krla daanna one, flot wage thwa thiynwa data type eken accept krna ewa unath puluwan)

tf.math.log(tf.cast(H, dtype=tf.float32))
```
