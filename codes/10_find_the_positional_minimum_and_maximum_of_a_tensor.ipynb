```python
 #finding the positional maximum and minimum
import tensorflow as tf

tf.random.set_seed(42)
tensor1 = tf.random.uniform(shape=[50])
tensor1
```

```python
# first lets see what is the maximum number of this tensor
tf.reduce_max(tensor1)
```

```python
#find the positional maximum/argmax is the method we use to find the ositional maximum
tf.argmax(tensor1)
```

```python
#to check if the argmx gave the right answer lets check the 42th position value:
tensor1[42]
```

```python
#check for equality
tensor1[tf.argmax(tensor1)] == tf.reduce_max(tensor1)
```

```python
#so now lets check positioanl minimum

#before that we have to check what is the minimum value of this tesnor

minValue = tf.reduce_min(tensor1)
```

```python
#now lets get the positional minimum with the relevant method :
tf.argmin(tensor1)
```

```python
#now lets check

tensor1[tf.argmin(tensor1)] == tf.reduce_min(tensor1)
tensor1[tf.argmin(tensor1)] == minValue
```
