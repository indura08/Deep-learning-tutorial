```python
#tensors and numpy:- tesnorflow interacts beautifuly wiht numpy arrays

import tensorflow as tf
import numpy as np

#create a tensor directly form a numpy array
tensor1 = tf.constant(np.array([3., 7., 10.]))
tensor1
```

```python
#convert ur tensor back to a nupy array
np.array(tensor1), type(np.array(tensor1))
```

```python
#another way to convert tesnor to a umpy array
tensor1.numpy(), type(tensor1.numpy())
```
